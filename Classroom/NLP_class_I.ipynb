```python
def remove_punc(text):
  for i in exclude:
    text = text.replace(i,'')
  return text
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def port_stem(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
w = 'walks walk walking walked'
```

```python
port_stem(w)
```
