## Step 1: Import Required Libraries

We import **Hugging Face Transformers** and **PyTorch**.

- `BertTokenizer` → converts raw text into numbers
- `BertForSequenceClassification` → BERT model with classification head
- `torch` → tensor operations and inference


```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
```

## Step 2: Load Pretrained Tokenizer

We load the **pretrained BERT tokenizer**.

### What happens internally?
- Loads WordPiece vocabulary
- Handles lowercasing
- Adds special tokens automatically

`bert-base-uncased` means:
- 12 encoder layers
- Lowercase English vocabulary


```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

## Step 3: Load BERT Model for Classification

We load BERT **with a classification head** on top.

### Internally:
- Base BERT encoder remains unchanged
- A new dense layer is added on top of `[CLS]` token
- `num_labels=2` → binary classification


```python
model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2
)
```

## Step 4: Input Text

This is the sentence we want BERT to understand.

BERT **does not read text directly**.
It must be converted into **input IDs** and **attention masks**.

```python
text = "I love studying deep learning"
```

## Step 5: Tokenization (MOST IMPORTANT STEP)

We convert raw text into BERT-readable format.

### What tokenizer does:
1. Splits text into WordPiece tokens
2. Adds `[CLS]` at start
3. Adds `[SEP]` at end
4. Converts tokens into IDs
5. Creates attention mask


```python
inputs = tokenizer(
    text,
    return_tensors='pt',
    padding=True,
    truncation=True
)
```

### Inspect Tokenized Output

- `input_ids` → numerical representation of tokens
- `attention_mask` → tells model which tokens are real


```python
inputs
```

## Step 6: Forward Pass Through BERT

We pass tokenized input into the model.

### Internally:
- Tokens → Embeddings
- Pass through 12 encoder layers
- `[CLS]` embedding extracted
- Classification head applied


```python
outputs = model(**inputs)
```

## Step 7: Understanding Model Output

`logits` are **raw scores**, not probabilities.

Shape:
- `(batch_size, num_labels)`

```python
logits = outputs.logits
logits
```

## Step 8: Prediction

We take the class with **highest score**.

`argmax` selects the predicted label.

```python
prediction = torch.argmax(logits, dim=1)
prediction.item()
```

## Final Summary

- BERT understands context using **bidirectional attention**
- `[CLS]` token represents the entire sentence
- Fine-tuning adds a small task-specific head
- Same BERT → multiple NLP tasks
