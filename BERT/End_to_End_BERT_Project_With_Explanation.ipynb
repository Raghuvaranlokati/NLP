# End-to-End BERT Project (With Full Explanation)

## Project Title: Sentiment Analysis using BERT

This notebook demonstrates a **complete real-world NLP pipeline** using BERT:

1. Load and understand data
2. Preprocess text
3. Tokenize using BERT tokenizer
4. Fine-tune BERT
5. Evaluate the model
6. Perform predictions

Every step is explained **line by line** in markdown cells.

## Step 1: Import Required Libraries

- `pandas` → data handling
- `torch` → deep learning backend
- `transformers` → BERT model & tokenizer
- `sklearn` → train-test split & metrics

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## Step 2: Create Dataset (Simulating Real Data)

In real projects, data comes from CSV, database, or API.
Here, we simulate a labeled dataset for **sentiment analysis**.

```python
data = {
    'text': [
        'I love this product',
        'This is the worst experience ever',
        'Absolutely fantastic service',
        'I am very disappointed',
        'Best purchase I have made',
        'Not worth the money'
    ],
    'label': [1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
df
```

## Step 3: Train-Test Split

We split data into:
- Training set → model learns patterns
- Test set → model evaluation


```python
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.3, random_state=42
)
```

## Step 4: Load BERT Tokenizer

Tokenizer converts text into:
- Input IDs
- Attention masks

These are required by BERT.

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

## Step 5: Tokenization Function

We define a reusable function to tokenize text data.

This ensures consistency during training and inference.

```python
def tokenize_texts(texts):
    return tokenizer(
        list(texts),
        padding=True,
        truncation=True,
        return_tensors='pt'
    )
```

## Step 6: Tokenize Train and Test Data

```python
train_inputs = tokenize_texts(X_train)
test_inputs = tokenize_texts(X_test)

train_labels = torch.tensor(y_train.values)
test_labels = torch.tensor(y_test.values)
```

## Step 7: Load BERT Model

We load BERT with a classification head.

- `num_labels=2` → binary sentiment classification

```python
model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2
)
```

## Step 8: Training Setup

We use:
- Adam optimizer
- Cross-entropy loss (built into model)


```python
optimizer = torch.optim.Adam(model.parameters(), lr=2e-5)
model.train()
```

## Step 9: Training Loop (Single Epoch for Demo)

For teaching purposes, we train for **one epoch**.

In real projects, multiple epochs are used.

```python
outputs = model(**train_inputs, labels=train_labels)
loss = outputs.loss
loss.backward()
optimizer.step()

print('Training loss:', loss.item())
```

## Step 10: Evaluation

We switch model to evaluation mode and disable gradients.

```python
model.eval()

with torch.no_grad():
    outputs = model(**test_inputs)
    predictions = torch.argmax(outputs.logits, dim=1)

accuracy = accuracy_score(test_labels, predictions)
accuracy
```

## Step 11: Real-Time Prediction (FINAL GOAL)

This is how BERT is used in **real applications**.

```python
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    prediction = torch.argmax(outputs.logits, dim=1)
    return 'Positive' if prediction.item() == 1 else 'Negative'

predict_sentiment('I really enjoyed this product')
```
